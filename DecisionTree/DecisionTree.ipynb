# Titanic

```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv("titanic.csv")
titanic
```

```python
Y_titan = titanic.Survived
Y_titan
```

```python
titan = titanic.drop(["Survived"], axis = "columns")
```

```python
titan
```

```python
dum = pd.get_dummies(titan.Sex)
dum
```

```python
merged = pd.concat([titan,dum], axis = "columns")
merged
```

```python
X_titan = merged.drop(["Sex", "female"], axis = "columns")
X_titan
```

```python
median = X_titan.Age.median()
median
```

```python
X_titan.Age = X_titan.fillna(median)
X_titan.Age
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_titan,Y_titan,test_size = 0.2)
x_train
```

```python
tree = tree.DecisionTreeClassifier()
```

```python
tree.fit(x_train, y_train)
```

```python
tree.predict(x_test)
```

```python
tree.score(x_test,y_test)
```
