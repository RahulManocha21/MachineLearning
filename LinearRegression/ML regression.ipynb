```python
 import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
ds = pd.read_csv('DS.csv')
```

```python
ds
```

```python
plt.scatter(ds.area,ds.prices, color = 'red', marker= "*")
plt.xlabel("area")
plt.ylabel("prices")
```

```python
reg = linear_model.LinearRegression()
reg.fit(ds[['area']][:4],ds.prices[:4])
```

```python
reg.predict([[1000]])
```

```python
p=reg.predict(ds[['area']][5:10])
```

```python
ds['prices'][5:10]= p
```

```python
ds
```

```python
ds.to_csv("DS.csv")
```

```python
plt.scatter(ds.area,ds.prices, color = 'red', marker= "*")
plt.xlabel("area")
plt.ylabel("prices")
plt.plot(ds.area,reg.predict(ds[['area']]), color = 'yellow')
```

# Multiple dependencies

```python
 import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import math as mt

hire = pd.read_csv('hiring.csv')
```

```python
hire
```

```python
from word2number import w2n
```

```python
a = []
for i in hire.experience:
    x= w2n.word_to_num(i)
    a.append(x)
```

```python
a
```

```python
hire['experience'] = a
```

```python
hire
```

```python
median = hire.test.median()
```

```python
median
```

```python
hire.test = hire.test.fillna(median)
```

```python
hire
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(hire[['experience','test','pi']],hire.salary)
```

```python
reg.predict([[2,10,10]])
```

```python
reg.score(hire[['experience','test','pi']],hire.salary)
```

# Dummies OneHot Encoding

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn import linear_model
```

```python
car = pd.read_csv("car.csv")
car
```

```python
dum = pd.get_dummies(car.Car)
dum
```

```python
merge = pd.concat([car,dum],axis="columns")
merge
```

```python
final = merge.drop(['Car','Mercedez Benz C class'], axis="columns")
final
```

```python
y = final.Price
```

```python
x = final.drop(['Price'],axis="columns")
```

```python
x
```

```python
y
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.predict([[57000,5,0,1]])
```

```python
reg.score(x,y)
```

```python
import joblib
```

```python
joblib.dump(reg,"model")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, train_size=0.8)
```

```python
x_test
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
reg.predict(x_test)
```
